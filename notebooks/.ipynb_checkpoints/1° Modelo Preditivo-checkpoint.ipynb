# Bibliotecas

```python
# Importação das bibliotecas de manipulação e gráficos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Bibliotecas sklearn para separação e pré-processamento
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.utils.class_weight import compute_class_weight

# Importação das métricas
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, roc_auc_score, precision_score, recall_score, accuracy_score, f1_score
from sklearn.model_selection import cross_val_score, KFold

# Importação dos Modelos
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Biblioteca para ignorar erros
import warnings
```

# Carregamento dos Dados

# Atributos Previsores e Alvo

```python
# Lista das colunas relevantes para o modelo
features = [
    'sexo', 'idade', 'motivo_internacao', 'doencas_base___1', 'doencas_base___2', 
    'doencas_base___3', 'doencas_base___4', 'doencas_base___5', 'doencas_base___6',
    'doencas_base___7', 'doencas_base___8', 'doencas_base___9', 'doencas_base___10',
    'imc', 'interface_vent_chegada'
]

# Variável target
target = ['desfechoUTI']
```

```python
# Seleção dos atributos (features) e da variável alvo (target)
x = df[features]
y = df[target].values.ravel()
```

# Separação de treino e teste

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5, stratify=y)
```

```python
print("Shape do conjunto de treino:", x_train.shape, y_train.shape)
print("Shape do conjunto de teste:", x_test.shape, y_test.shape)
```

# Escalonamento e OneHotEncoder

```python
# Definindo as colunas numéricas e categóricas
num_features = [
    'idade', 'imc'
]

cat_features = [
    'sexo', 'motivo_internacao', 'doencas_base___1', 'doencas_base___2', 
    'doencas_base___3', 'doencas_base___4', 'doencas_base___5', 'doencas_base___6',
    'doencas_base___7', 'doencas_base___8', 'doencas_base___9', 'doencas_base___10',
    'interface_vent_chegada'
]
```

```python
# Identificação de colunas binárias
binary_cols = [col for col in cat_features if set(df[col].unique()).issubset({0, 1})]
true_cat_cols = [col for col in cat_features if col not in binary_cols]
```

```python
# Aplicação do StandardScaler nas variáveis numéricas
scaler = StandardScaler()
x_train_num = scaler.fit_transform(x_train[num_features])
x_test_num = scaler.transform(x_test[num_features])
```

```python
# Aplicação do OneHotEncoder nas variáveis categóricas não binárias
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
x_train_true_cat = encoder.fit_transform(x_train[true_cat_cols])
x_test_true_cat = encoder.transform(x_test[true_cat_cols])

# Mantendo as binárias originais
x_train_binary = x_train[binary_cols].values
x_test_binary = x_test[binary_cols].values
```

```python
# Combinar features numéricas e categóricas
x_train_processed = np.hstack((x_train_num, x_train_true_cat, x_train_binary))
x_test_processed = np.hstack((x_test_num, x_test_true_cat, x_test_binary))
```

# Treinamento e Seleção do Modelo

```python
# Definação dos modelos a serem testados
models = {
    'Logistic Regression': LogisticRegression(random_state=5),
    'XGBoost': XGBClassifier(random_state=5, eval_metric='logloss'),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(random_state=5, probability=True),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=5),
    'Random Forest': RandomForestClassifier(random_state=5)
}

# Definição de parâmetros para testar os modelos
parameters = {
    # Regressão Logística
    'Logistic Regression': {
        'C': [0.1, 1, 10, 100],
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear', 'saga'],
        'max_iter': [100, 200, 500, 1000],
        'tol': [0.0001],
        'class_weight': [{0: 3, 1:1}]
    },

    # XGBoost
    'XGBoost': {
        'n_estimators': [100, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1],
        'scale_pos_weight': [0.33, 0.5, 1]
    },

    # Naive Bayes
    'Naive Bayes': {},

    # SVM
    'SVM': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto'],
        'class_weight': [{0: 3, 1:1}]
    },

    # KNN
    'KNN': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance']
    },

    # Árvore de Decisão
    'Decision Tree': {
        'max_depth': [None, 3, 5, 7],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': [{0: 3, 1:1}]
    },

    # Random Forest
    'Random Forest': {
        'n_estimators': [100, 200],
        'max_features': ['sqrt', 'log2', 0.2],
        'max_depth': [None, 3, 5, 7],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': [{0: 3, 1:1}]
    }
}
```

```python
best_model = None
best_score = 0
best_params = None
results = {}

for model_name in models:
    print(f"\n--- Treinando {model_name} ---")
    clf = GridSearchCV(
        models[model_name],
        parameters[model_name],
        cv=5,
        scoring=['roc_auc', 'f1', 'accuracy'],
        refit='roc_auc',
        n_jobs=-1
    )
    clf.fit(x_train_processed, y_train)
    
    # Armazena resultados
    results[model_name] = {
        'best_score': clf.best_score_,
        'best_params': clf.best_params_,
        'best_estimator': clf.best_estimator_
    }
    
    print(f"Melhor ROC-AUC: {clf.best_score_:.3f}")
    print(f"Melhores parâmetros: {clf.best_params_}")
    
    if clf.best_score_ > best_score:
        best_model = clf.best_estimator_
        best_score = clf.best_score_
        best_params = clf.best_params_

print("\n" + "="*50)
print(f"Melhor modelo geral: {best_model.__class__.__name__}")
print(f"ROC-AUC médio: {best_score:.3f}")
print(f"Melhores parâmetros do modelo vencedor:\n{best_params}")
```

# Avaliação do Modelo

```python
# Previsões finais
y_pred = best_model.predict(x_test_processed)
print("Relatório de Classificação:\n", classification_report(y_test, y_pred))
```

```python
# Calcule a matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels = [0, 1])

# Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Óbito', 'Não óbito'],
            yticklabels=['Óbito', 'Não óbito'])
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```

```python
# Curva ROC
y_pred_proba = best_model.predict_proba(x_test_processed)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falso Positivo')
plt.ylabel('Taxa de Verdadeiro Positivo')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
```

```python
# Validação cruzada para o melhor modelo
kf = KFold(n_splits=5, shuffle=True, random_state=5)

cv_scores = cross_val_score(best_model, x_train_processed, y_train, cv=kf, scoring='roc_auc')

print("\nValidação Cruzada - ROC-AUC Scores:", cv_scores)
print(f"Média do ROC-AUC na Validação Cruzada: {cv_scores.mean():.3f}")
print(f"Desvio Padrão do ROC-AUC na Validação Cruzada: {cv_scores.std():.3f}")
```

```python
# Lista para armazenar resultados
metrics_list = []

# Avaliação para cada modelo
for model_name, result in results.items():
    model = result['best_estimator']
    y_pred = model.predict(x_test_processed)
    y_prob = model.predict_proba(x_test_processed)[:, 1]

    # Matriz de confusão para especificidade
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    # Cálculo das medidas
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    specificity = tn / (tn + fp)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    metrics_list.append({
        'Modelo': model_name,
        'Precisão': precision,
        'Sensibilidade': recall,
        'Acurácia': accuracy,
        'F1-Score': f1,
        'AUC-ROC': auc,
    })

import warnings
warnings.filterwarnings("ignore", module='joblib')

# Criar tabela final
metrics_df = pd.DataFrame(metrics_list)
print(metrics_df)
```

# Análise de Features

```python
# Nomes das features numéricas (já escaladas)
num_feature_names = num_features

# Nomes das features categóricas (após OneHotEncoding)
cat_feature_names = encoder.get_feature_names_out(true_cat_cols)

# Nomes das features binárias (mantidas originais)
binary_feature_names = binary_cols

# Combinar todos os nomes na ordem correta
all_feature_names = np.concatenate([num_feature_names, cat_feature_names, binary_feature_names])
```

### Random Forest

```python
# Obter o modelo Random Forest
rf_model = results['Random Forest']['best_estimator']

# Extrair importância das features
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

# Plotar as 15 features mais importantes
plt.figure(figsize=(12, 8))
plt.title("Top 15 Features - Random Forest")
plt.barh(range(15), importances[indices][:15][::-1], align='center')
plt.yticks(range(15), [all_feature_names[i] for i in indices[:15]][::-1])
plt.xlabel('Importância Relativa')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
